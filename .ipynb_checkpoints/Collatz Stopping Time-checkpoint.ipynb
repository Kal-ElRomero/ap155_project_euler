```python
limit = 10000
```

```python
def T(co):
    Colls = [co]
    cn = co
    while cn != 1:
        if cn % 2 == 0:
            cn = cn//2
        else:
            cn = 3*cn + 1
        Colls.append(cn)
    T_len = len(Colls)
    return T_len
```

```python
import matplotlib.pyplot as plot

def Colplot(Ns, Ts):
    plot.xlabel("N", fontsize=11)
    plot.ylabel("T(N)", fontsize=11)
    plot.plot(Ns, Ts, 'bo', markersize=3)
    plot.show()
```

```python
Ts = [T(N) for N in range(1,limit+1)]
Ns = [N for N in range(1,limit+1)]

Colplot(Ns, Ts)
```
